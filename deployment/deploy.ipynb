```python
# =========================================================================
# SageMaker Endpoint Deployment
# =========================================================================

from sagemaker.sklearn.model import SKLearnModel
import sagemaker

role = sagemaker.get_execution_role()
sess = sagemaker.Session()

# Assuming model artifacts are tarred and uploaded to this S3 path
s3_model_path = "s3://my-bucket/models/churn-ensemble-v1.tar.gz"

model = SKLearnModel(
    model_data=s3_model_path,
    role=role,
    entry_point="inference.py",
    source_dir="../deployment", # folder where inference.py + requirements.txt exist
    framework_version="1.2-1",  # Using SKLearn container as base
    py_version="py3"
)

predictor = model.deploy(
    instance_type="ml.c4.2xlarge",
    initial_instance_count=1,
    endpoint_name="prod-churn-ensemble-endpoint",
    update_endpoint=True
)

print(f"Endpoint deployed successfully: {predictor.endpoint_name}")
```
